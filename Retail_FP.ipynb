#Data Preprocessing

```python
import numpy as np
transactions = []
for sublist in store_data.values.tolist():
  clean_sublist = [item for item in sublist if item is not np.nan]
  transactions.append(clean_sublist)
```

```python
transactions
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

#FP-Growth Algorithm Implementation

```python
!pip install mlxtend --upgrade --no-deps
```

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets = fpgrowth(df, min_support=0.0045, use_colnames=True)

frequent_itemsets 
```

#Association Rules
Association rule mining finds interesting associations and relationships among large sets of data items. This rule shows how frequently a itemset occurs in a transaction.

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
```

#Comparison of Performance between Apriori and FP-Growth Algorithm

```python
from mlxtend.frequent_patterns import apriori

%timeit -n 100 -r 10 apriori(df, min_support=0.0045)
#3.02 ms  118 s per loop (mean  std. dev. of 10 runs, 100 loops each)
```
